## The main factors which impacts the house price in Kansas City 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
import mpl_toolkits
%matplotlib inline
```

```python
data = pd.read_csv("./kc_house_data.csv")
```

```python
snsplot = sns.jointplot(x = "lat", y = "long", kind="scatter", data=data)
```

```python
x = data['sqft_living']
y = data['price']
house_avg_price = round(y.mean(), 20)
size = y / house_avg_price * 20
plt.figure(figsize=(20, 10))
plt.title('The relationship between the house living area and the house price')
fig = plt.scatter(x,y, s=size, alpha=0.5, label='house living area compared to the average house living area in KC')
# maximum likelihood estimation
x = sm.add_constant(x)
model = sm.OLS(y, x).fit()
predicts = model.predict()
x = data['sqft_living']
plt.plot(x, predicts, color='red', label='maximum likelihood estimation')

# legend
plt.legend(loc='upper left')
plt.xlabel('house living area')
plt.ylabel('price')

plt.show()
```

```python
x = data['sqft_lot']
y = data['price']
house_avg_price = round(y.mean(), 20)
size = y / house_avg_price * 20
plt.figure(figsize=(20, 10))
plt.title('The relationship between the house lot area and the house price')
fig = plt.scatter(x, y, s=size, alpha=0.5)
plt.legend(handles = [fig],labels=['house lot area compared to the average house lot area in KC'],loc='best')

plt.xlabel('house lot area')
plt.ylabel('price')
plt.show()
```
